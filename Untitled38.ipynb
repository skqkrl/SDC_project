```python
# 1) 연식 범주화
af['건축연도대'] = pd.cut(af['건축년도'], bins=[1900, 1989, 1999, 2009, 2019, 2030], labels=['~89', '90s', '00s', '10s', '20s'])

# 2) 면적 범주화
af['면적대'] = pd.cut(af['전용면적(㎡)'], bins=[0, 30, 50, 70, 90, 200], labels=['~30', '30-50', '50-70', '70-90', '90+'])

# 3) 유사 조건 그룹핑
group_cols = ['구', '주택유형', '건축연도대', '면적대']
af['유사조건그룹'] = af[group_cols].astype(str).agg('_'.join, axis=1)

# 4) 각 그룹의 평균 보증금과 표준편차 계산
group_stats = af.groupby('유사조건그룹')['보증금(만원)'].agg(['mean', 'std']).reset_index()
group_stats.columns = ['유사조건그룹', '유사군_평균보증금', '유사군_표준편차']

# 5) 원본과 merge
af = af.merge(group_stats, on='유사조건그룹', how='left')
```

```python
# Z-score 방식: 평균 ± 2표준편차 밖이면 이상치
af['보증금_zscore'] = (af['보증금(만원)'] - af['유사군_평균보증금']) / af['유사군_표준편차']
af['보증금_이상치여부'] = af['보증금_zscore'].abs() > 2
```

```python
#1단계 : 구 단위로 데이터 정리
import pandas as pd

# 예시로 불러온 데이터프레임: df

# 1) 필요한 컬럼 정리
df = df[['구', '전용면적(㎡)', '보증금(만원)', '계약년월', '계약일', '건축년도', '계약구분', '도로명', '종전계약 보증금(만원)']]

# 2) 계약연월 컬럼 만들기
df['계약연월'] = df['계약년월'].astype(str)

# 3) 건축년도 구간화
df['건축연도대'] = pd.cut(df['건축년도'], bins=[1900, 1989, 1999, 2009, 2019, 2031],
                        labels=['~1989', '1990s', '2000s', '2010s', '2020s'])

# 4) 면적 구간화
df['면적대'] = pd.cut(df['전용면적(㎡)'], bins=[0, 30, 50, 70, 100, 999],
                     labels=['~30', '30-50', '50-70', '70-100', '100+'])
```

```python
### 2. 구별 핵심 이상 탐지 지표 만들기

### 지표 1: 보증금 이상치 (유사 조건 그룹 대비)
# 유사 조건 그룹 생성
df['조건그룹'] = df[['구', '건축연도대', '면적대']].astype(str).agg('_'.join, axis=1)

# 그룹별 평균/표준편차
group_stats = df.groupby('조건그룹')['보증금(만원)'].agg(['mean', 'std']).reset_index()
group_stats.columns = ['조건그룹', '보증금_평균', '보증금_표준편차']

# 병합 후 z-score 계산
df = df.merge(group_stats, on='조건그룹', how='left')
df['보증금_zscore'] = (df['보증금(만원)'] - df['보증금_평균']) / df['보증금_표준편차']
df['보증금_이상치'] = df['보증금_zscore'].abs() > 2  # Z-score 2 초과이면 이상치
```

```python
#지표 2: 동일 주소 + 월 거래수 ≥ 3 → 사기 의심 클러스터
df['도로명_연월'] = df['도로명'] + '_' + df['계약연월']
주소_빈도 = df['도로명_연월'].value_counts().reset_index()
주소_빈도.columns = ['도로명_연월', '주소_해당월_거래수']

# 병합
df = df.merge(주소_빈도, on='도로명_연월', how='left')
df['거래_집중의심'] = df['주소_해당월_거래수'] >= 3
```

```python
#### 지표 3: 노후건물 중 이상 보증금 비율
df['노후건물'] = df['건축년도'] < 1990
df['노후_이상보증'] = df['노후건물'] & df['보증금_이상치']
```

```python

#3단계 : 구 단위로 집계구별_지표

구별_지표= df.groupby('구').agg({
    '보증금_이상치': 'sum',
    '거래_집중의심': 'sum',
    '노후_이상보증': 'sum',
    '도로명': 'count'  # 총 거래 수
}).reset_index()

# 비율 계산
구별_지표['보증금_이상비율'] = 구별_지표['보증금_이상치'] / 구별_지표['도로명']
구별_지표['거래집중비율'] = 구별_지표['거래_집중의심'] / 구별_지표['도로명']
구별_지표['노후이상비율'] = 구별_지표['노후_이상보증'] / 구별_지표['도로명']
```

```python
# 구별 종합 위험 점수 계산 (정규화)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
score_cols = ['보증금_이상비율', '거래집중비율', '노후이상비율']
구별_지표[[f'{col}_정규화' for col in score_cols]] = scaler.fit_transform(구별_지표[score_cols])

# 종합 점수 = 평균
구별_지표['위험스코어'] = 구별_지표[[f'{col}_정규화' for col in score_cols]].mean(axis=1)

# 위험 상위 구 보기
상위구 = 구별_지표.sort_values('위험스코어', ascending=False)
```

```python
# 유사 조건 그룹 생성
ff['조건그룹'] = ff[['구', '건축연도대', '면적대']].astype(str).agg('_'.join, axis=1)

# 그룹별 평균/표준편차
group_stats = ff.groupby('조건그룹')['보증금(만원)'].agg(['mean', 'std']).reset_index()
group_stats.columns = ['조건그룹', '보증금_평균', '보증금_표준편차']

# 병합 후 z-score 계산
ff = ff.merge(group_stats, on='조건그룹', how='left')
ff['보증금_zscore'] = (ff['보증금(만원)'] - ff['보증금_평균']) / ff['보증금_표준편차']
ff['보증금_이상치'] = ff['보증금_zscore'].abs() > 2  # Z-score 2 초과이면 이상치
```

```python
ff['단가'] = ff['보증금(만원)'] / ff['전용면적(㎡)']

# 자치구별 IQR 기반 이상치 탐지
def detect_iqr_outliers(group):
    q1 = group['단가'].quantile(0.25)
    q3 = group['단가'].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    group['이상거래'] = ~group['단가'].between(lower, upper)
    return group

ff = ff.groupby('구').apply(detect_iqr_outliers)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def analyze_risk(ef):
    # -------------------
    # 1. 전처리
    # -------------------
    ef = ef[['구', '계약면적(㎡)', '보증금(만원)', '계약년월', '계약일', '건축년도', '계약구분', '도로명', '종전계약 보증금(만원)']]
    ef['계약연월'] = ef['계약년월'].astype(str)

    ef['건축연도대'] = pd.cut(ef['건축년도'], bins=[1900, 1989, 1999, 2009, 2019, 2031],
                            labels=['~1989', '1990s', '2000s', '2010s', '2020s'])
    ef['면적대'] = pd.cut(ef['계약면적(㎡)'], bins=[0, 30, 50, 70, 100, 999],
                         labels=['~30', '30-50', '50-70', '70-100', '100+'])

    # -------------------
    # 2. 지표 계산
    # -------------------
    # 2-1. 보증금 이상치 (Z-score)
    ef['조건그룹'] = ef[['구', '건축연도대', '면적대']].astype(str).agg('_'.join, axis=1)
    group_stats = ef.groupby('조건그룹')['보증금(만원)'].agg(['mean', 'std']).reset_index()
    group_stats.columns = ['조건그룹', '보증금_평균', '보증금_표준편차']
    ef = ef.merge(group_stats, on='조건그룹', how='left')
    ef['보증금_zscore'] = (ef['보증금(만원)'] - ef['보증금_평균']) / ef['보증금_표준편차']
    ef['보증금_이상치'] = ef['보증금_zscore'].abs() > 2

    # 2-2. 거래 집중 클러스터
    ef['도로명_연월'] = ef['도로명'] + '_' + ef['계약연월']
    addr_month_freq = ef['도로명_연월'].value_counts().reset_index()
    addr_month_freq.columns = ['도로명_연월', '주소_해당월_거래수']
    ef = ef.merge(addr_month_freq, on='도로명_연월', how='left')
    ef['거래_집중의심'] = ef['주소_해당월_거래수'] >= 3

    # 2-3. 노후 건물 + 이상 보증금
    ef['노후건물'] = ef['건축년도'] < 1990
    ef['노후_이상보증'] = ef['노후건물'] & ef['보증금_이상치']

    # 2-4. 신규 건물 + 이상 보증금 (신규 건물: 2020년 이후)
    ef['신규건물'] = ef['건축년도'] >= 2020
    ef['신규_이상보증'] = ef['신규건물'] & ef['보증금_이상치']

    # -------------------
    # 3. 구별 집계
    # -------------------
    구별_지표 = ef.groupby('구').agg({
        '보증금_이상치': 'sum',
        '거래_집중의심': 'sum',
        '노후_이상보증': 'sum',
        '신규_이상보증': 'sum',
        '도로명': 'count'
    }).reset_index()

    구별_지표['보증금_이상비율'] = 구별_지표['보증금_이상치'] / 구별_지표['도로명']
    구별_지표['거래집중비율'] = 구별_지표['거래_집중의심'] / 구별_지표['도로명']
    구별_지표['노후이상비율'] = 구별_지표['노후_이상보증'] / 구별_지표['도로명']
    구별_지표['신규이상비율'] = 구별_지표['신규_이상보증'] / 구별_지표['도로명']

    # -------------------
    # 4. 위험스코어 계산
    # -------------------
    score_cols = ['보증금_이상비율', '거래집중비율', '노후이상비율', '신규이상비율']
    scaler = MinMaxScaler()
    구별_지표[[f'{col}_정규화' for col in score_cols]] = scaler.fit_transform(구별_지표[score_cols])
    구별_지표['위험스코어'] = 구별_지표[[f'{col}_정규화' for col in score_cols]].mean(axis=1)

    return 구별_지표.sort_values('위험스코어', ascending=False)
```
