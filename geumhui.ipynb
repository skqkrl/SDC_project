```python
df_2023 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2023년.csv",  encoding='cp949', header=15)
df_2024 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2024년.csv",  encoding='cp949', header=15)
```

```python
df_deal2023 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2023년.csv",  encoding='cp949', header=15)
df_deal2024 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2024년.csv",  encoding='cp949', header=15)
```

```python
print("2023년 단독 다가구 전월세")
display(df_2023.head())
print("2024년 단독 다가구 전월세")
display(df_2024.head())
```

```python
# 단독 다가구 전월세
print("2023년 단독 다가구 전월세")
print(df_2023.info())
print("2024년 단독 다가구 전월세")
print(df_2024.info())
```

```python
# 데이터 모든 컬럼 동일함으로 '번지', '도로조건' 컬럼 제거
# '전월세구분'에서 전세만 추출
drop_columns = ['번지','도로조건']

# 컬럼 제거 및 필터링
def filtered_df(df):
    # 1. 번지, 도로 저건 제거
    df_filtered = df.drop(columns = drop_columns, errors = 'ignore')

    # 2. 전월세구분 '전세'인 행만 추출
    df_filtered = df_filtered[df_filtered['전월세구분'] == '전세']

    return df_filtered.reset_index(drop = True)


# 3. 각 연도별 데이터셋 처리
df_2023_cleaned = filtered_df(df_2023)
df_2024_cleaned = filtered_df(df_2024)

print("2023년 단독 다가구 전세")
display(df_2023_cleaned.head())

print("2024년 단독 다가구 전세")
display(df_2024_cleaned.head())
```

```python
# 각 년도별 컬럼 결측값
display(pd.concat(
    [
        df_2023_cleaned.isnull().sum(),
        df_2024_cleaned.isnull().sum()
    ],
    axis=1,
    keys=['2023', '2024']
).reset_index().rename(columns={'index': 'column'}))
```

```python
# 보증금(만원), 월세금(만원) 데이터 타입 int로 변경

# 전처리 함수 정의
def clean_rent_columns(df):
    for col in ['보증금(만원)', '월세금(만원)']:
        df[col] = (
            df[col]
            .astype(str)               # 문자열로 변환
            .str.replace(',', '')      # 쉼표 제거
            .str.strip()               # 공백 제거
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')  # 숫자로 변환, 오류시 NaN
        df[col] = df[col].fillna(0).astype(int)            # NaN은 0으로 대체 후 int
    return df

# 데이터프레임별로 처리
df_2023_cleaned = clean_rent_columns(df_2023_cleaned)
df_2024_cleaned = clean_rent_columns(df_2024_cleaned)

# 결과 확인
print("2023년 단독 다가구 전세")
display(df_2023_cleaned.head())

print("2024년 단독 다가구 전세")
display(df_2024_cleaned.head())
```

```python
# 계약 년월 datetime으로 타입 변경
df_2023_cleaned['계약년월'] = pd.to_datetime(df_2023_cleaned['계약년월'].astype(str), format='%Y%m')
df_2024_cleaned['계약년월'] = pd.to_datetime(df_2024_cleaned['계약년월'].astype(str), format='%Y%m')
```

```python
df_2023_cleaned = df_2023_cleaned[df_2023_cleaned['계약년월']>='2023-06-01']
df_2023_cleaned['계약년월'].min()
```

```python
# 구별 거래건수 집계
gu_counts_23 = df_2023_cleaned['구'].value_counts()
gu_counts_24 = df_2023_cleaned['구'].value_counts()

# 합치기 후 결측값 처리 및 정수형 변환
combined = gu_counts_23.add(gu_counts_24, fill_value=0).astype(int)

# 시리즈 → 데이터프레임 변환
combined_df = combined.reset_index()
combined_df.columns = ['구', '거래건수']

# 거래건수 기준 정렬
combined_df = combined_df.sort_values(by='거래건수', ascending=False)

# 결과 출력
combined_df
```
