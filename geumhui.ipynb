```python
df_deal2021 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2021년.csv",  encoding='cp949', header=15)
df_deal2022 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2022년.csv",  encoding='cp949', header=15)
df_deal2023 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2023년.csv",  encoding='cp949', header=15)
df_deal2024 = pd.read_csv("C:/Users/rmagm/OneDrive/바탕 화면/vscode/서울특별시 단독,다가구 전월세 데이터/단독다가구(전월세)_실거래가_2024년.csv",  encoding='cp949', header=15)
```

```python
# 단독 다가구 전월세
print("2021년 단독 다가구 전월세")
print(df_2021.info())
print("2022년 단독 다가구 전월세")
print(df_2022.info())
print("2023년 단독 다가구 전월세")
print(df_2023.info())
print("2024년 단독 다가구 전월세")
print(df_2024.info())
```

```python
# 데이터 모든 컬럼 동일함으로 '번지', '도로조건' 컬럼 제거
# '전월세구분'에서 전세만 추출
drop_columns = ['번지','도로조건']

# 컬럼 제거 및 필터링
def filtered_df(df):
    # 1. 번지, 도로 저건 제거
    df_filtered = df.drop(columns = drop_columns, errors = 'ignore')

    # 2. 전월세구분 '전세'인 행만 추출
    df_filtered = df_filtered[df_filtered['전월세구분'] == '전세']

    return df_filtered.reset_index(drop = True)


# 3. 각 연도별 데이터셋 처리
df_2021_cleaned = filtered_df(df_2021)
df_2022_cleaned = filtered_df(df_2022)
df_2023_cleaned = filtered_df(df_2023)
df_2024_cleaned = filtered_df(df_2024)

print("2021년 단독 다가구 전세")
display(df_2021_cleaned)

print("2022년 단독 다가구 전세")
display(df_2022_cleaned)

print("2023년 단독 다가구 전세")
display(df_2023_cleaned)

print("2024년 단독 다가구 전세")
display(df_2024_cleaned)
```

```python
# 각 년도별 컬럼 결측값
display(pd.concat(
    [
        df_2021_cleaned.isnull().sum(),
        df_2022_cleaned.isnull().sum(),
        df_2023_cleaned.isnull().sum(),
        df_2024_cleaned.isnull().sum()
    ],
    axis=1,
    keys=['2021', '2022', '2023', '2024']
).reset_index().rename(columns={'index': 'column'}))
```

```python
# 보증금(만원), 월세금(만원) 데이터 타입 int로 변경

# 전처리 함수 정의
def clean_rent_columns(df):
    for col in ['보증금(만원)', '월세금(만원)']:
        df[col] = (
            df[col]
            .astype(str)               # 문자열로 변환
            .str.replace(',', '')      # 쉼표 제거
            .str.strip()               # 공백 제거
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')  # 숫자로 변환, 오류시 NaN
        df[col] = df[col].fillna(0).astype(int)            # NaN은 0으로 대체 후 int
    return df

# 데이터프레임별로 처리
df_2021_cleaned = clean_rent_columns(df_2021_cleaned)
df_2022_cleaned = clean_rent_columns(df_2022_cleaned)
df_2023_cleaned = clean_rent_columns(df_2023_cleaned)
df_2024_cleaned = clean_rent_columns(df_2024_cleaned)

# 결과 확인
print("2021년 단독 다가구 전세")
display(df_2021_cleaned)

print("2022년 단독 다가구 전세")
display(df_2022_cleaned)

print("2023년 단독 다가구 전세")
display(df_2023_cleaned)

print("2024년 단독 다가구 전세")
display(df_2024_cleaned)
```

```python
print("2021년 단독 다가구 전세")
display(df_2021_cleaned['계약구분'].value_counts().reset_index())
print("2022년 단독 다가구 전세")
display(df_2022_cleaned['계약구분'].value_counts().reset_index())
print("2023년 단독 다가구 전세")
display(df_2023_cleaned['계약구분'].value_counts().reset_index())
print("2024년 단독 다가구 전세")
display(df_2024_cleaned['계약구분'].value_counts().reset_index())
```
