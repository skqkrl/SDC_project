```python
# 데이터 모든 컬럼 동일함으로 '번지', '도로조건' 컬럼 제거

drop_columns = ['번지','도로조건']

# 컬럼 제거 및 필터링
def filtered_df(df):
    # 1. 번지, 도로 조건 제거
    df_filtered = df.drop(columns = drop_columns, errors = 'ignore')

    # 2. 전월세구분 '전세'인 행만 추출
    df_filtered = df_filtered[df_filtered['전월세구분'] == '전세']

    return df_filtered.reset_index(drop = True)


# 3. 각 연도별 데이터셋 처리
df_2023_cleaned = filtered_df(df_2023)
df_2024_cleaned = filtered_df(df_2024)

print("2023년 단독 다가구 전세")
display(df_2023_cleaned.head())

print("2024년 단독 다가구 전세")
display(df_2024_cleaned.head())
```

```python
# 계약년월 + 계약일 => "계약시기" 컬럼 생성

def prepare_contract_data(df, 기준일='2023-06-01'):
    df = df.copy()
    
    # 계약 시기 컬럼 생성: 'YYYYMM' + '일' 조합
    df['계약 시기'] = pd.to_datetime(
        df['계약년월'].astype(str).str[:4] + '-' +
        df['계약년월'].astype(str).str[4:6] + '-' +
        df['계약일'].astype(str).str.zfill(2),
        format='%Y-%m-%d',
        errors='coerce'
    )
    
    # 기준일 이후로 필터링
    df = df[df['계약 시기'] >= pd.to_datetime(기준일)]
    
    return df
```

```python
multi_household_2023 = prepare_contract_data(df_2023_cleaned)
multi_household_2024 = prepare_contract_data(df_2024_cleaned)
```

```python
# 보증금(만원), 월세금(만원) 데이터 타입 int로 변경

# 전처리 함수 정의
def clean_rent_columns(df):
    for col in ['보증금(만원)', '월세금(만원)','종전계약 보증금(만원)','종전계약 월세(만원)']:
        df[col] = (
            df[col]
            .astype(str)               # 문자열로 변환
            .str.replace(',', '')      # 쉼표 제거
            .str.strip()               # 공백 제거
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')  # 숫자로 변환, 오류시 NaN
        df[col] = df[col].fillna(0).astype(int)            # NaN은 0으로 대체 후 int
    return df

# 데이터프레임별로 처리
multi_household_2023 = clean_rent_columns(multi_household_2023)
multi_household_2024 = clean_rent_columns(multi_household_2024)

# 결과 확인
print("2023년 단독 다가구 전세")
display(multi_household_2023.head())

print("2024년 단독 다가구 전세")
display(multi_household_2024.head())
```

```python
def extract_contract_months(row):
    try:
        if pd.isna(row['계약기간']) or '~' not in row['계약기간']:
            return None
        start_str, end_str = row['계약기간'].split('~')
        start_date = pd.to_datetime(start_str + '01', format='%Y%m%d', errors='coerce')
        end_date = pd.to_datetime(end_str + '01', format='%Y%m%d', errors='coerce')

        if pd.isna(start_date) or pd.isna(end_date):
            return None

        return (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
    except:
        return None

multi_household['계약개월수'] = multi_household.apply(extract_contract_months, axis=1)
```

### 동 컬럼 생성

```python
# '동' 컬럼 추출 함수 정의
def extract_dong(sigungu_address):
    if pd.isna(sigungu_address):
        return None
    
    parts = sigungu_address.split('구') # '구' 기준으로 분할
    if len(parts) > 1:
        # '구 ' 다음의 문자열이 '동' 이름일 가능성이 높음
        dong_name = parts[1].strip()
        return dong_name
    return None # '구'가 없거나 패턴이 맞지 않는 경우

# '시군구' 컬럼에서 '동' 정보 추출하여 새로운 '동' 컬럼 생성
multi_household['동'] = multi_household['시군구'].apply(extract_dong)

# 결과 확인
print(multi_household[['시군구', '동']].head(10))
```

### 갱신요구권 사용 컬럼값

```python
print("--- 처리 전 데이터프레임 ---")
print(multi_household)
print("\n'계약개월수' 결측값 개수 (처리 전):", multi_household['계약개월수'].isna().sum())
print("'갱신요구권 사용' 결측값 개수 (처리 전):", multi_household['갱신요구권 사용'].isna().sum())

# --- 2. '갱신요구권 사용' 컬럼 처리 (도메인 지식 반영하여 세분화) ---

# 2-1. '갱신요구권 사용'의 원래 결측 여부를 나타내는 플래그 생성 (옵션이지만 권장)
# 이 플래그는 나중에 범주화된 값이 원래 결측이었는지 여부를 알려줍니다.
multi_household['is_갱신요구권_사용_originally_missing'] = multi_household['갱신요구권 사용'].isna().astype(int)


# 2-2. '계약구분'이 '신규'인 경우 '갱신요구권 사용'을 '해당없음'으로 설정
multi_household.loc[multi_household['계약구분'] == '신규', '갱신요구권 사용'] = '해당없음'

# 2-3. '계약구분'이 '갱신'인데 '갱신요구권 사용'이 결측(NaN)이거나 특정 미사용 값인 경우 '사용하지 않음'으로 설정
# 여기서는 np.nan, '-', '' (빈 문자열)을 미사용으로 간주합니다.
multi_household.loc[
    (multi_household['계약구분'] == '갱신') &
    (multi_household['갱신요구권 사용'].isin([np.nan, '-', ''])), # NaN 또는 '-' 또는 빈 문자열
    '갱신요구권 사용'
] = '미사용'

# 2-4. 만약 '미사용'이라는 값이 이미 명시적으로 있다면, 그대로 둡니다. (위 로직에 의해 영향받지 않음)
#      '사용' 또는 'O' 값도 그대로 유지됩니다.

print("\n--- '갱신요구권 사용' 도메인 처리 후 ---")
print(multi_household)
print("\n'갱신요구권 사용' 결측값 개수 (중간 처리 후):", multi_household['갱신요구권 사용'].isna().sum())
```
