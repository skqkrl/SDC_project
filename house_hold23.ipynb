## 1. 행정구역별 거주 주택유형

```python
admit21_df = pd.read_excel("C:/Users/rmagm/OneDrive/바탕 화면/vscode/SDC_project/house_hold23,24/house_hold21_admit.xlsx")
admit22_df = pd.read_excel("C:/Users/rmagm/OneDrive/바탕 화면/vscode/SDC_project/house_hold23,24/house_hold22_admit.xlsx")
admit23_df = pd.read_excel("C:/Users/rmagm/OneDrive/바탕 화면/vscode/SDC_project/house_hold23,24/house_hold23_admit.xlsx")
```

```python
print("2021년 행정구역별 주택유형")
display(admit21_df.head())
print("2022년 행정구역별 주택유형")
display(admit22_df.head())
print("2023년 행정구역별 주택유형")
display(admit23_df.head())
```

```python
# 컬럼명 변경
def rename_columns(df):
    return df.rename(columns = {'기타':"비주거용 건물 내 주택",'다가구원룸':"오피스텔",'단독다가구':"단독주택",
                             "비주택기타":"기타", "아파트": "아파트","연립다세대":"연립주택","오피스텔":"다세대주택"})
# 각 연도별 DataFrame에 적용
admit21_df = rename_columns(admit21_df)
admit22_df = rename_columns(admit22_df)
admit23_df = rename_columns(admit23_df)
```

```python
print("2021년 행정구역별 주택유형")
display(admit21_df.head())
print("2022년 행정구역별 주택유형")
display(admit22_df.head())
print("2023년 행정구역별 주택유형")
display(admit23_df.head())
```

```python
# 단독주택과 다세대 주택 "단독다세대" 컬럼으로 합침
def multi_household(df):
    # '단독다세대' 컬럼 생성
    df['단독다가구'] = df['단독주택'] + df['다세대주택']

    # 기존 컬럼 제거
    df = df.drop(['단독주택', '다세대주택'], axis = 1)

    return df

admit21 = multi_household(admit21_df)
admit22 = multi_household(admit22_df)
admit23 = multi_household(admit23_df)
```

```python
print("2021년 행정구역별 주택유형")
display(admit21.head())
print("2022년 행정구역별 주택유형")
display(admit22.head())
print("2023년 행정구역별 주택유형")
display(admit23.head())
```

```python
admit21.to_csv("household_admit21.csv", index=False, encoding='utf-8')
admit22.to_csv("household_admit22.csv", index=False, encoding='utf-8')
admit23.to_csv("household_admit23.csv", index=False, encoding='utf-8')
```
