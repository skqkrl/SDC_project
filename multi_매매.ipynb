```python
df_23 = pd.read_csv('C:/Users/rmagm/OneDrive/바탕 화면/vscode/SDC_project/multi_household/단독다가구(매매)_실거래가_2023년.csv',encoding='cp949', header=15)
df_24 = pd.read_csv('C:/Users/rmagm/OneDrive/바탕 화면/vscode/SDC_project/multi_household/단독다가구(매매)_실거래가_2024년.csv', encoding='cp949', header=15)
```

```python
multi_deal = pd.concat([df_23, df_24])
```

```python
multi_deal['계약시기'] = pd.to_datetime(multi_deal['계약년월'].astype(str) + multi_deal['계약일'].astype(str))
```

```python
multi_deal = multi_deal[multi_deal['계약시기']>='2023-06-01']
```

```python
multi_deal['구'] = multi_deal['시군구'].str.extract(r'(\S+구)')
multi_deal.isnull().sum()
```

```python
multi_deal['거래금액(만원)'] = multi_deal['거래금액(만원)'].astype(str).str.replace(',', '', regex=False)
multi_deal['거래금액(만원)'] = pd.to_numeric(multi_deal['거래금액(만원)']).astype(int)
```
