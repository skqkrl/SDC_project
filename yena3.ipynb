```python
# 1) 필요한 컬럼 정리
df = total_df[['주택유형','구','동', '전용면적(㎡)', '전세금', '계약시기', '건축년도', '계약구분', '종전계약 보증금(만원)','전세가율']]

# 2) 계약연월 컬럼 만들기
df['계약시기'] = df['계약시기'].astype(str)

# 3) 건축년도 구간화
df['건축연도대'] = pd.cut(df['건축년도'], bins=[1900, 1989, 1999, 2009, 2019, 2031],
                        labels=['~1989', '1990s', '2000s', '2010s', '2020s'])

# 4) 면적 구간화
df['면적대'] = pd.cut(df['전용면적(㎡)'], bins=[0, 30, 50, 70, 100, 999],
                     labels=['~30', '30-50', '50-70', '70-100', '100+'])
```

```python
# 유사 조건 그룹 생성
df['조건그룹'] = df[['구', '건축연도대', '면적대']].astype(str).agg('_'.join, axis=1)

# 그룹별 평균/표준편차
group_stats = df.groupby('조건그룹')['전세금'].agg(['mean', 'std']).reset_index()
group_stats.columns = ['조건그룹', '보증금_평균', '보증금_표준편차']

# 병합 후 z-score 계산
df = df.merge(group_stats, on='조건그룹', how='left')
df['보증금_zscore'] = (df['전세금'] - df['보증금_평균']) / df['보증금_표준편차']
df['보증금_이상치'] = df['보증금_zscore'].abs() > 2  # Z-score 2 초과이면 이상치
```

```python
df['도로명_연월'] = df['구'] + '_' + df['계약시기']
주소_빈도 = df['도로명_연월'].value_counts().reset_index()
주소_빈도.columns = ['도로명_연월', '주소_해당월_거래수']

# 병합
df = df.merge(주소_빈도, on='도로명_연월', how='left')
df['거래_집중의심'] = df['주소_해당월_거래수'] >= 3
```

```python
df['노후건물'] = df['건축년도'] < 1990
df['노후_이상보증'] = df['노후건물'] & df['보증금_이상치']
```

```python
# 1. 기본 지표 집계
구별_지표 = df.groupby('구').agg({
    '보증금_이상치': 'sum',
    '거래_집중의심': 'sum',
    '노후_이상보증': 'sum',
    '동': 'count',  # 총 거래 수
    '전세가율': 'mean'  # 전세가율은 평균으로 사용
}).reset_index()

# 2. 비율 계산
구별_지표['보증금_이상비율'] = 구별_지표['보증금_이상치'] / 구별_지표['동']
구별_지표['거래집중비율'] = 구별_지표['거래_집중의심'] / 구별_지표['동']
구별_지표['노후이상비율'] = 구별_지표['노후_이상보증'] / 구별_지표['동']
```

```python
# 각 비율 및 전세가율을 정규화 (Z-score 방식 또는 Min-Max 가능)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
score_cols = ['보증금_이상비율', '거래집중비율', '노후이상비율', '전세가율']
구별_지표[[f'{col}_score' for col in score_cols]] = scaler.fit_transform(구별_지표[score_cols])

# 4. 위험 점수 = 단순 평균 (가중치 주고 싶으면 각자 곱해주기)
구별_지표['위험스코어'] = 구별_지표[[f'{col}_score' for col in score_cols]].mean(axis=1)

# 5. 정렬
구별_지표 = 구별_지표.sort_values('위험스코어', ascending=False)

# 위험 상위 구 보기
상위구 = 구별_지표.sort_values('위험스코어', ascending=False).head(5)
```
