## 전세사기 개정법 이후의 연립다세대 전세 데이터

```python
df= pd.read_csv('/content/drive/MyDrive/틈만나면/연립다세대/townhouse.csv')
```

```python
def eda_summary(df, name="DataFrame"):
    print(f"\n====== {name} ======")
    print(f"Shape: {df.shape}")
    print("\nData Types:")
    print(df.dtypes)
    print("\nMissing Values:")
    print(df.isnull().sum())
    print(f"\nDuplicated Rows: {df.duplicated().sum()}")
    print("=====================\n")

eda_summary(df, name="연립다세대 주택 전세 데이터")
```

```python
af=df.groupby(['시군구','건축년도'])['보증금(만원)'].mean()
af
```

```python
region_counts= df['시군구'].value_counts()
```

```python
top_regions = region_counts.sort_values(ascending=False).head(30)
top_regions_df = top_regions.reset_index()
top_regions_df.columns = ['시군구', '거래건수']
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

plt.figure(figsize=(10, 6))
plt.barh(top_regions_df['시군구'], top_regions_df['거래건수'])
plt.xlabel('거래건수')
plt.ylabel('시군구')
plt.title('시군구별 거래건수 상위 20개')
plt.gca().invert_yaxis()  # 높은 값이 위로 오게
plt.show()
```

```python
# 보증금(만원) 과 전용면적의 상관관계
correlation = df['보증금(만원)'].corr(df['전용면적(㎡)'])
print(f"보증금과 평수의 상관계수: {correlation:.3f}")
```

```python
df['보증금(만원)'].describe()
```

```python
df['보증금(만원)'].head(5)
```

```python
df_clean = df[df['보증금(만원)'] < 100000]
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
plt.figure(figsize=(8,6))
plt.scatter(df_clean['전용면적(㎡)'], df_clean['보증금(만원)'], alpha=0.5)
plt.xlabel('전용면적(㎡)')
plt.ylabel('보증금(만원)')
plt.title('보증금 vs 평수')
plt.show()
```

###전용면적이 클수록 보증금은 대체로 상승하는 경향은 있으나, 동일 면적 내에서도 입지, 연식, 주택유형 등에 따라 보증금 편차가 크다.

```python
# 연식별 보증금 차이 확인
correlation2 = df_clean['보증금(만원)'].corr(df_clean['건축년도'])
print(f"보증금과 건축년도의 상관계수: {correlation2:.3f}")
```

```python
df_clean.groupby('건축년도')['보증금(만원)'].mean().plot(figsize=(10,6))
plt.ylabel("평균 보증금(만원)")
plt.title("건축년도별 평균 보증금 변화")
plt.show()
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
plt.figure(figsize=(8,6))
plt.scatter(df_clean['건축년도'], df_clean['보증금(만원)'], alpha=0.5)
plt.xlabel('건축년도')
plt.ylabel('보증금(만원)')
plt.title('건축년도 vs 보증금')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 예시 데이터 준비
# df : 이미 주어진 테이블 그대로 있다고 가정
housing_types = ['기타','다가구원룸','단독다가구','비주택기타','아파트','연립다세대','오피스텔']

# 비율로 변환
df_ratio = houseage2023.copy()
df_ratio_total = df_ratio[housing_types].sum(axis=1)
for col in housing_types:
    df_ratio[col] = df_ratio[col] / df_ratio_total

# 스택 막대그래프 그리기
df_ratio.set_index('가구주연령대')[housing_types].plot(kind='bar', stacked=True, figsize=(12,8))

plt.title("가구주 연령대별 주택유형 구성 (비율 기준)")
plt.ylabel("비율")
plt.xlabel("가구주 연령대")
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.show()
```
