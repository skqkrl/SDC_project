```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 데이터 생성
data = {
    '구분': ['전세사기피해자등 결정', '긴급한 경·공매 유예'],
    '가결': [30400, 997],
    '부결': [8268, 67]
}

df = pd.DataFrame(data)
df.set_index('구분', inplace=True)

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for ax, (index, row) in zip(axes, df.iterrows()):
    labels = ['가결', '부결']
    sizes = row.values
    colors = ['skyblue', 'salmon']

    # 퍼센트와 건수를 동시에 표시하는 함수
    def make_autopct(sizes):
        def my_autopct(pct):
            total = sum(sizes)
            count = int(round(pct * total / 100.0))
            return f'{pct:.1f}%\n({count}건)'
        return my_autopct

    wedges, texts, autotexts = ax.pie(
        sizes, labels=labels, autopct=make_autopct(sizes), colors=colors, startangle=90
    )
    ax.set_title(f"{index}")

plt.suptitle("가결 vs 부결 파이차트")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 데이터 입력
data = {
    '지역': [
        '서울', '경기', '대전', '인천', '부산',
        '전남', '대구', '경북', '전북', '광주',
        '세종', '경남', '강원', '충남', '충북',
        '울산', '제주'
    ],
    '건수': [
        8334, 6657, 3569, 3341, 3328,
        1001, 669, 531, 443, 440,
        424, 389, 360, 351, 282,
        185, 96
    ]
}

df = pd.DataFrame(data)
df = df.sort_values(by='건수', ascending=False).reset_index(drop=True)

# 색상 설정
highlight_colors = ['#A82810','#F67B50', '#FBC490', '#FBAA60', '#FFB347']
gray_colors = ['#d3d3d3'] * (len(df) - len(highlight_colors))
colors = highlight_colors + gray_colors

# 총 건수 계산
total = df['건수'].sum()


# 도넛형 파이차트
plt.figure(figsize=(12, 12))

wedges, texts, autotexts = plt.pie(
    df['건수'],
    labels=df['지역'],
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    wedgeprops=dict(width=0.4),
    colors=colors,
    labeldistance=1.1
)

# 라벨 겹침 방지: 흰 배경
for text in texts:
    text.set_fontsize(12)
    text.set_fontweight('bold')
    text.set_bbox(dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.3'))

# 퍼센트 글씨 설정
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')
    autotext.set_color('black')

# 중앙 텍스트
plt.text(0, 0, f'지역별 전세사기 건수\n{total:,}건', ha='center', va='center', fontsize=18, weight='bold')



plt.tight_layout()
plt.show()
```
