## 계약기간 개월수를 구하기=>


```python

DF= townhouse.copy()
DF
```

```python
def calculate_month_difference(contract_period_str):
    if not isinstance(contract_period_str, str) or '~' not in contract_period_str:
            return None
    try:
        start_date_str, end_date_str = contract_period_str.split('~')

        # 시작 년도와 월 추출
        start_year = int(start_date_str[:4])
        start_month = int(start_date_str[4:])

        # 종료 년도와 월 추출
        end_year = int(end_date_str[:4])
        end_month = int(end_date_str[4:])

        # 개월수 차이 계산
        # (종료년도 - 시작년도) * 12 + (종료월 - 시작월)
        month_difference = (end_year - start_year) * 12 + (end_month - start_month)

        return month_difference
    except Exception as e:
        return f"Error: {e}"
```

```python
DF.loc[:, '계약개월수'] = DF['계약기간'].apply(calculate_month_difference)
```

```python
filtered_df = DF[(DF['계약개월수'] >= 1) & (DF['계약개월수'] < 60)].copy()
```

```python
# 조건: 계약구분이 '신규'이고 계약 시기가 결측이 아닌 경우
condition = (filtered_df['계약구분'] == '신규') & (filtered_df['계약시기'].notna())

# 이 조건을 만족하면서 계약개월수가 NaN인 경우에만 24로 채움
filtered_df.loc[condition & (filtered_df['계약개월수'].isna()), '계약개월수'] = 24

# 계약 개월 수가 결측(NaN)이고, 갱신 + 갱신요구권 사용 == '사용'인 경우 → 24개월로 채움
condition1 = (
    filtered_df['계약개월수'].isna() & # '계약개월수'가 NaN인 경우를 확인
    (filtered_df['계약구분'] == '갱신') &
    (filtered_df['갱신요구권 사용'].isin(['사용', 'O']))
)
filtered_df.loc[condition1, '계약개월수'] = 24
```

```python
# 갱신인 경우 계약 개월 수 nan 값 처리 못함 -> 모델링에서 쓰는 것으로 컬럼 처리
# 계약개월수 NaN 여부 플래그 추가
filtered_df['is_계약개월수_missing'] = filtered_df['계약개월수'].isna().astype(int)
# 남은 NaN을 -1로 대체
filtered_df['계약개월수'] = filtered_df['계약개월수'].fillna(-1)
```

```python
print("--- 처리 전 데이터프레임 ---")
print(filtered_df)
print("\n'계약개월수' 결측값 개수 (처리 전):", filtered_df['계약개월수'].isna().sum())
print("'갱신요구권 사용' 결측값 개수 (처리 전):", filtered_df['갱신요구권 사용'].isna().sum())

# --- 2. '갱신요구권 사용' 컬럼 처리 (도메인 지식 반영하여 세분화) ---

# 2-1. '갱신요구권 사용'의 원래 결측 여부를 나타내는 플래그 생성 (옵션이지만 권장)
# 이 플래그는 나중에 범주화된 값이 원래 결측이었는지 여부를 알려줍니다.
filtered_df['is_갱신요구권_사용_originally_missing'] = filtered_df['갱신요구권 사용'].isna().astype(int)


# 2-2. '계약구분'이 '신규'인 경우 '갱신요구권 사용'을 '해당없음'으로 설정
filtered_df.loc[filtered_df['계약구분'] == '신규', '갱신요구권 사용'] = '해당없음'

# 2-3. '계약구분'이 '갱신'인데 '갱신요구권 사용'이 결측(NaN)이거나 특정 미사용 값인 경우 '사용하지 않음'으로 설정
# 여기서는 np.nan, '-', '' (빈 문자열)을 미사용으로 간주합니다.
filtered_df.loc[
    (filtered_df['계약구분'] == '갱신') &
    (filtered_df['갱신요구권 사용'].isin([np.nan, '-', ''])), # NaN 또는 '-' 또는 빈 문자열
    '갱신요구권 사용'
] = '미사용'

# 2-4. 만약 '미사용'이라는 값이 이미 명시적으로 있다면, 그대로 둡니다. (위 로직에 의해 영향받지 않음)
#      '사용' 또는 'O' 값도 그대로 유지됩니다.

print("\n--- '갱신요구권 사용' 도메인 처리 후 ---")
print(filtered_df)
print("\n'갱신요구권 사용' 결측값 개수 (중간 처리 후):", filtered_df['갱신요구권 사용'].isna().sum())
```
