```python
for df in [df_2021_buy, df_2022_buy, df_2023_buy,df_2024_buy]:
    df['연도'] = df['계약년월'] // 100
```

```python
df_buy_all= pd.concat([df_2021_buy, df_2022_buy, df_2023_buy, df_2024_buy], ignore_index=True)
df_buy_all.head(5)
```

```python
df_buy_all['계약년월'].min(), df_buy_all['계약년월'].max()
```

```python
df_buy_all= df_buy_all.sort_values(by='계약년월',ascending=True)
df_buy_all['거래금액(만원)'] = df_buy_all['거래금액(만원)'].astype(str).str.replace(',', '')
df_buy_all['거래금액(만원)'] = pd.to_numeric(df_buy_all['거래금액(만원)'])
df_buy_all['월'] = df_buy_all['계약년월'] % 100
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import koreanize_matplotlib
plt.figure(figsize=(10,6))
sns.boxplot(x='연도', y='거래금액(만원)', data=df_buy_all, palette='Set2')
plt.title('연도별 거래금액 분포 비교')
plt.ylabel('거래금액 (만원)')
plt.show()
```

```python
# 연도-월별 거래건수 집계
monthly_count = df_buy_all.groupby(['연도', '월']).size().reset_index(name='거래건수')
monthly_pivot = monthly_count.pivot(index='월', columns='연도', values='거래건수').sort_index()
monthly_pivot
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

plt.figure(figsize=(10,6))

# 라인그래프 출력
for year in monthly_pivot.columns:
    plt.plot(monthly_pivot.index, monthly_pivot[year], marker='o', label=str(year))

plt.title('월별 연립다세대 거래건수 (연도별)')
plt.xlabel('월')
plt.ylabel('거래건수')
plt.xticks(range(1,13))  # 1월~12월
plt.legend(title='연도')
plt.grid(True)
plt.show()
```

```python
for df in [df_2021_rent, df_2022_rent, df_2023_rent, df_2024_rent]:
    df['연도'] = df['계약년월'] // 100
```

```python
df_rent_all= pd.concat([df_2021_rent, df_2022_rent, df_2023_rent, df_2024_rent], ignore_index=True)
df_rent_all.head(5)
```

```python
df_rent_all.columns
```

```python
df_rent_all= df_rent_all.sort_values(by='계약년월',ascending=True)
df_rent_all['보증금(만원)'] = df_rent_all['보증금(만원)'].astype(str).str.replace(',', '')
df_rent_all['보증금(만원)'] = pd.to_numeric(df_rent_all['보증금(만원)'])
df_rent_all['월'] = df_rent_all['계약년월'] % 100
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import koreanize_matplotlib
plt.figure(figsize=(10,6))
sns.boxplot(x='연도', y='보증금(만원)', data=df_rent_all, palette='Set2')
plt.title('연도별 전세 거래금액 분포 비교')
plt.ylabel('전세 거래금액 (만원)')
plt.show()
```

```python
# 연도-월별 거래건수 집계
monthly_count = df_rent_all.groupby(['연도', '월']).size().reset_index(name='거래건수')
monthly_pivot = monthly_count.pivot(index='월', columns='연도', values='거래건수').sort_index()
monthly_pivot
```

```python
# ④ 전세만 필터링
df_jeonse = df_rent_all[df_rent_all['전월세구분'] == '전세'].copy()

# ⑤ 계약년월에서 월만 추출
df_jeonse['월'] = df_jeonse['계약년월'] % 100

# ⑥ 월별 거래건수 집계
monthly_count = df_jeonse.groupby(['연도', '월']).size().reset_index(name='거래건수')

# ⑦ Pivot 변환 (그래프 그리기 용이하도록)
monthly_pivot = monthly_count.pivot(index='월', columns='연도', values='거래건수').sort_index()

# ⑧ 시각화
plt.figure(figsize=(10,6))

for year in monthly_pivot.columns:
    plt.plot(monthly_pivot.index, monthly_pivot[year], marker='o', label=str(year))

plt.title('4개년 월별 전세 거래건수')
plt.xlabel('월')
plt.ylabel('전세 거래건수')
plt.xticks(range(1,13))
plt.legend(title='연도')
plt.grid(True)
plt.show()
```

```python
# 종전계약 보증금 값 존재 여부 컬럼 추가
df_rent_all['종전보증금존재'] = df_rent_all['종전계약 보증금(만원)'].notna()

# 계약구분별 종전계약 보증금 존재 비율 확인
result = df_rent_all.groupby('계약구분')['종전보증금존재'].mean().reset_index()
result['존재비율(%)'] = result['종전보증금존재'] * 100
#갱신은 종전계약 보증금이 거의 항상 있음, 신규는 거의 없음 → 이 불일치가 의심 거래 탐지에 도움.
result
```

```python
# 연도별, 시군구별 전세 거래건수 집계
grouped = df_jeonse.groupby(['연도', '시군구']).size().reset_index(name='거래건수')

# 각 연도별로 전세 거래건수 가장 많은 시군구 추출
top_region_by_year = grouped.sort_values(['연도', '거래건수'], ascending=[True, False]).groupby('연도').head(10)

top_region_by_year
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 예시 데이터 대체용 (삭제해도 됨)
# top_region_by_year = pd.DataFrame({
#     '연도': [2023, 2024],
#     '시군구': ['서울 강서구', '인천 서구'],
#     '거래건수': [3124, 2812]
# })

# 필터링
top_recent = top_region_by_year[top_region_by_year['연도'].isin([2023, 2024])]
print(top_recent)  # 비어있지 않은지 확인

# 시각화
plt.figure(figsize=(6, 4))
plt.bar(top_recent['연도'].astype(str), top_recent['거래건수'], color='skyblue')

for i, row in top_recent.iterrows():
    plt.text(row['연도'], row['거래건수'] + 10, row['시군구'], ha='center', fontsize=10)

plt.title('2023~2024 전세 거래량 상위 지역')
plt.xlabel('연도')
plt.ylabel('거래건수')
plt.tight_layout()
plt.show()
```
