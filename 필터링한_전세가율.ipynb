# 조건 추가해서 매매가 평균 구하기

```python
DF['건축년도'].isnull().sum()
```

```python
bins_year = [1900, 1989, 1999, 2009, 2019, 2031]
labels_year = ['~1989', '1990s', '2000s', '2010s', '2020s']
DF['건축년도대'] = pd.cut(DF['건축년도'], bins=bins_year, labels=labels_year, right=False)
```

```python
DF['전용면적(㎡)'].isnull().sum()
```

```python
bins_area = [0, 30, 50, 70, 100, 999]
labels_area = ['~30', '30-50', '50-70', '70-100', '100+']
DF['면적대'] = pd.cut(DF['전용면적(㎡)'], bins=bins_area, labels=labels_area, right=False)
DF
```

```python
group_cols = ['구','건축년도대', '면적대']
buy_mean=DF.groupby(group_cols, observed=True)['매매가'].mean().reset_index()
buy_mean.columns = group_cols + ['그룹_평균_매매가(만원)']
```

```python
buy_mean
```

# 조건별 평균 매매가

```python
af= pd.read_csv('/content/drive/MyDrive/틈만나면/연립다세대/townhouse.csv')
af.head()
```

```python
af.columns
```

```python
af['구'] = af['시군구'].str.extract(r'([가-힣]+구)')
af['전세가'] = af['보증금(만원)'].replace(',', '', regex=True).astype(int)
```

```python
af.isnull().sum()
```

```python
bins_area = [0, 30, 50, 70, 100, 999]
labels_area = ['~30', '30-50', '50-70', '70-100', '100+']
af['면적대'] = pd.cut(af['전용면적(㎡)'], bins=bins_area, labels=labels_area, right=False)
```

```python
af['면적대'].isnull().sum()
```

```python
bins_year = [1900, 1989, 1999, 2009, 2019, 2031]
labels_year = ['~1989', '1990s', '2000s', '2010s', '2020s']
af['건축년도대'] = pd.cut(af['건축년도'], bins=bins_year, labels=labels_year, right=False)
```

```python
multi_townhouse = pd.merge(af,buy_mean, on=group_cols, how='left')
```

```python
multi_townhouse['전세가율'] = (multi_townhouse['보증금(만원)'] / multi_townhouse['그룹_평균_매매가(만원)']) * 100
```
